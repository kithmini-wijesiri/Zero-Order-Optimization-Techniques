**Minimizing a quadratic function and the curse of dimensionality**


Consider the simple quadratic function
ùëî(ùë§)=ùë§ùëáùë§ 
whose minimum is always at the origin regardless of the input dimension  ùëÅ .
Create a range of these quadratics for input dimension  ùëÅ=1  to  ùëÅ=100 , sample the input space of each  ùëÉ=100  times randomly on the hypercube  [‚àí1,1]√ó[‚àí1,1]√ó‚Ä¶√ó[‚àí1,1]  (this hypercube has  ùëÅ  sides), and plot the minimum value attained for each quadratic against the input dimenionsion  ùëÅ .

```python
import numpy as np
np.random.seed(0)
```

```python
# define the function g using one line of code
g = lambda w: np.dot(w.T,w)
```

```python
# implement the function xmins, which takes one argument P, the number of samples from the hypercube,
# and returns the minimum values of g among these P samples 
# for dimension N = 1, 2, ..., 100.
def xmins(P):
  xmins = []
  for N in range(1,101):
    x = np.random.uniform(-1.0,1.0, (P,N))
    evals = np.array([g(w) for w in x])
    xmins.append(np.min(evals))
  return np.array(xmins)
```

```python
# get the results from P = 100, 1000, and 10000
xmins_100 = xmins(100)
xmins_1000 = xmins(1000)
xmins_10000 = xmins(10000)
```

```python
# plot the above results
import matplotlib.pyplot as plt

plt.plot(np.arange(1,101), xmins_100);
plt.plot(np.arange(1,101), xmins_1000);
plt.plot(np.arange(1,101), xmins_10000);
plt.xlabel("N");
plt.ylabel("min");
plt.legend(["P=100", "P=1000", "P=10000"]);
```
